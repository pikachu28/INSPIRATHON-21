```python
from tensorflow.keras.models import load_model
import numpy as np
import pickle
import joblib
#from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#loaded_model = pickle.load(open('cb_sgd_final.sav', 'rb'))
loaded_model = joblib.load('cb_sgd_final.sav')
tokenizer = pickle.load(open('tokenizer1.pkl', 'rb'))
```

```python
text = ["You are full of drama", "Nice!!", "Loved it!", "Boring", " bitch get up off me ", "momma said no pussy cats inside my doghouse"]

#tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(text)

word_index = tokenizer.word_index
sequence = tokenizer.texts_to_sequences(text)
#sequence = np.array(sequence)
```
